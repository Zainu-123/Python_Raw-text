```python
#Importing packages
import pandas as pd
from pprint import pprint as pp
import re
```

```python
# Opening file and reading line by line.Have done utf-encoding on the text and strip function to remove all leading and trailing 
# white spaces
with open('conv.txt', encoding="utf-8") as file:
    lines = file.readlines()   
    lines = [line.rstrip() for line in lines] 
print(lines)
```

```python
# To remove spaces after each line.
lines = list(filter(None, lines))
print(lines)
```

```python
#To print all characters as a list
characters =  []
for line in lines:
    character = line.split(':')[0]
    characters.append(character)
print(characters)
```

```python
#To print the unique characters by making a set of previous output and back to a list
unique_characters = list(set(characters))
print(unique_characters)
```

```python
print("The number of unique characters in the sample conversation is : %s"%len(unique_characters))
```

```python
#Creating a dictionary of names:words spoken
char_dict = {}
for name in unique_characters:
    char_dict[name] = []

for line in lines:
    name = line.split(':')[0]
    words = line.split(':')[1].strip().split(' ')
    for word in words:
        char_dict[name].append(word)
pp(char_dict)
```

```python
#Removing any word which is just a sequence of special characters.
for name in char_dict:
    reqd_list = []
    for word in char_dict[name]:
        reqd_word = re.sub('[?.,!;]{1,}','',word)
        reqd_list.append(reqd_word)
    char_dict[name] = reqd_list
pp(char_dict)
```

```python
for name in char_dict:
    del_list = []
    for word in char_dict[name]:
        if re.search('[A-Za-z]{1,}',word) is None:
            del_list.append(word)
    reqd_list = char_dict[name]
    for word in char_dict[name]:
        if word in del_list:
            reqd_list.remove(word)
    char_dict[name]=reqd_list
    char_dict[name]=list(set(char_dict[name]))
pp(char_dict)
```

```python
for name in char_dict:
    file = open('Output Files/'+name+".txt","w")
    for word in char_dict[name]:
        file.write(word+"\n")
    file.close()
```
